```python
#balancing the imbalance data
from sklearn.utils import resample
df_maj = parkinson_data[(parkinson_data['status']==1)] 
df_min = parkinson_data[(parkinson_data['status']==0)] 
df_min_upsampled = resample(df_min, replace=True,n_samples= 147, random_state=40) 
df_upsampled = pd.concat([df_min_upsampled, df_maj])
```

```python
df=df_upsampled.drop(columns=['name'],axis=1)
df.columns
```

```python
#dividing into test and train data
from sklearn.preprocessing import StandardScaler

X =df_upsampled.drop(columns=['name','status'], axis=1)
y = df_upsampled['status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25,random_state=102)
```

```python
i=1
plt.figure(figsize=(40,40))
for c in X_train.columns[:22]:
    plt.subplot(10,5,i)
    plt.title(f" distplot of {c}",fontsize=16)
    plt.yticks(fontsize=12)
    sns.distplot(x=X_train[c],color="green")
    i+=1
plt.show()
```

```python
#standardization
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
#svm
from sklearn.metrics import accuracy_score 
from sklearn import svm
model = svm.SVC(kernel='linear')
model.fit(X_train, y_train)
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
